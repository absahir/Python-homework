```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('/Users/anukritisharma/Desktop/Fintech/Jupyter-Wprkspace/menu_data.csv')
sales_filepath = Path('/Users/anukritisharma/Desktop/Fintech/Jupyter-Wprkspace/sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []
menu_count=0
sales_count=0
# @TODO: Read in the menu data into the menu list

with open(menu_filepath,'r') as menu_csv:
    #Pass in the csv file to csv.reader function
    menu_reader= csv.reader(menu_csv, delimiter=',')
    #print(type(csv_reader))
    #read header first
    csv_header=next(menu_reader)
    
    for row in menu_reader:   
        menu.append(row)
        menu_count += 1

# print(menu[2][1])
# print(menu_count)


#Reading in Sales Data from the sales_data.csv file

with open(sales_filepath,'r') as sales_csv:
    #Pass in the csv file to csv.reader function
    sales_reader= csv.reader(sales_csv, delimiter=',')
    #print(type(csv_reader))
    #read header first
    csv_header=next(sales_reader)
    
    for row in sales_reader:   
        sales.append(row)
        sales_count += 1

# print(sales[2][3])
# print(sales_count)


# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}
count=0     #The total quantity for each ramen type
revenue=0   #The total revenue for each ramen type
cogs=0      #The total cost of goods sold for each ramen type
profit=0    #The total profit for each ramen type

# Initialize a row counter variable
row_count = 0


## @TODO: Loop over every row in the sales list object

for item in sales:
    
    
    # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
    # @TODO: Initialize sales data variables
    
    item_id=item[0]
    date=item[1]
    ccard=item[2]
    quantity=item[3]
    sales_item=item[4]
    
    

    if sales_item not in report:
        
        report[sales_item]={"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}
        
for item in sales:
    quantity=item[3]
    sales_item=item[4]
    
    if sales_item in report:
        report[sales_item]["01-count"] += int(quantity)
        
for sales_index in sales:
    
    sales_item = sales_index[4]
    quantity = sales_index[3]
    
    for menu_row in menu:
        menu_item = menu_row[0]
        menu_price = menu_row[3]
        menu_cost = menu_row[4]
        
        if sales_item == menu_item:
            report[menu_item]["02-revenue"] += (int(menu_price) * int(quantity))
            report[menu_item]["03-cogs"] += (int(menu_cost) * int(quantity))
        elif sales_item != menu_item:
            continue
 
## calculate profit using revenue and cogs
for item, valuedict in report.items():
     
    for key in valuedict:
        
        cogs = report[item]["03-cogs"]
        revenue = report[item]["02-revenue"]
        
        if key == "04-profit":
            report[item][key] = revenue - cogs
        else:
            continue 

# output file with metrics
output_path = Path("PyRamen.txt")

with open(output_path, 'w') as file:
    file.write("This is the financial report for PyRamen.\n")
    for key in report:
        file.write(f"{key} {report[key]} \n")






        # @TODO: Calculate profit of each item in the menu data
```
